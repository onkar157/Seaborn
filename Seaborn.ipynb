# Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
```

```python
df.head(5)
```

```python
plt.figure(figsize=(8,6))
df['size'].value_counts().plot(kind='pie', autopct = '%1.1f%%')
```

```python
#scatter plot
plt.figure(figsize=(8,6), facecolor='lightgreen')
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

### Univariate --> one variable data visualisation

```python
#stirp plot
plt.style.use("seaborn-darkgrid")
plt.figure(figsize=(7,5), facecolor='lightpink')
sns.stripplot(y= 'tip', data=df, jitter=True)
plt.ylabel('Tip($)')
plt.show()
```

```python
#Grouping stripplot
plt.figure(figsize=(8,6), facecolor= 'lightgreen')
sns.stripplot(x='size', y='tip',data=df, jitter = True)
plt.xlabel=('Size')
plt.ylabel=('Tip-$')
plt.show()
```

```python
#Creating Boxplot
plt.figure(figsize=(8,6), facecolor= 'lightgreen')
sns.boxplot(x='size', y='tip',data=df)
plt.xlabel=('Size')
plt.ylabel=('Tip-$')
plt.show()
```

```python
plt.figure(figsize=(8,6), facecolor= 'lightgreen')
sns.boxplot(x='time', y='tip',data=df)
plt.xlabel=('time')
plt.ylabel=('Tip-$')
plt.show()
```

```python
df[df['time']=='Lunch'] ['tip'].quantile([0.25,0.50,0.75])
```

```python
sns.stripplot(x='day',y='tip', hue='sex', data=df, size=5, jitter=True)
plt.xlabel=('day')
plt.ylabel=('tip')
plt.show()
```

```python
sns.boxplot(x='day',y='tip',  data=df)
plt.xlabel=('day')
plt.ylabel=('tip')
plt.show()
```

```python
sns.boxplot(x='smoker',y='tip',  data=df)
plt.xlabel=('smoker')
plt.ylabel=('tip')
plt.show()
```

```python
sns.boxplot(x='sex',y='tip',  data=df)
plt.xlabel=('sex')
plt.ylabel=('tip')
plt.show()
```
